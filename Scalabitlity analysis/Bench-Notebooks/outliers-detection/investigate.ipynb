```python
import pandas as pd
```

```python
df1 = pd.read_csv("result_mpd.csv")
df2 = pd.read_csv("result_ptrail.csv")
```

```python
speed1 = df1.sort_values(by=["id","Timestamp"]).reset_index()[["id","Timestamp","geometry","speed"]]
speed1.to_csv("result_mpd.csv")
```

```python
speed2 = df2.sort_values(by=["traj_id","DateTime"]).reset_index()[["traj_id","DateTime","lat","lon","Speed"]]
speed2.to_csv("result_ptrail.csv")
```
